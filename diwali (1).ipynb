```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
df=pd.read_csv("B:\\programs and project\\data analysis project\\practice\\Python_Diwali_Sales_Analysis\\Diwali Sales Data.csv", encoding="unicode_escape")
df.head(10)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
df.head()
```

```python
#remove_null_values
df.dropna(inplace=True)
```

```python
#checking_for_ull_values
pd.isnull(df).sum()
```

```python
df.duplicated().sum()
```

```python
#drop_duplicates
df.drop_duplicates(inplace=True)
df.shape
df.dtypes
```

```python
#change_dtypes

df.dtypes
```

```python
df['Gender']=df['Gender'].replace({'M':'Male'})
df['Gender']=df['Gender'].replace({'F':'Female'})
df['Gender']
```

```python
df.describe()
df.dtypes
```

```python
plt.figure(figsize=(12,6))
sns.set_style("dark")
sns.set_context("notebook")
gg=df.groupby("State")['Orders'].sum().sort_values(ascending=False).reset_index()
k=sns.barplot(x="State", y="Orders",data=gg, width=0.8)
plt.xticks(rotation=45)
plt.title("state and order reltionship")
plt.xlabel("Orders")
plt.ylabel("State")
for bars in k.containers:
    k.bar_label(bars)

plt.show()
```

```python
Grouped=df.groupby(['State'])
print(Grouped['Orders'].sum())
```

```python
gen=df.groupby(['Gender'])['Orders'].sum().reset_index()
g=sns.barplot(x="Gender",y="Orders",data=gen, width=0.5)
sns.set_style("dark")
plt.title("man-women order ratio")
plt.ylabel('orders form market')
plt.xticks(rotation=45)
plt.xlabel('Gender male-female')
for bars in g.containers:
    g.bar_label(bars)
plt.show()
```

```python
occ=df.groupby(['Occupation'])['Orders'].sum().sort_values(ascending=False).reset_index()
ll=sns.barplot(x="Occupation", y="Orders", data=occ ,width=1)
plt.xticks(rotation=45, ha="right")
sns.set(rc={'figure.figsize':(20,10)})
plt.title("occupation-order relationship")
for bars in ll.containers:
    ll.bar_label(bars)
plt.show()
```

```python
df['Age_category']=df["Age"].apply(lambda x:'Teen' if 0<=x<=25 else 'Middle Aged' if 26<=x<=49 else "Old")
df.head(20)
```

```python
aoa=df.groupby("Age_category")[["Orders","Amount"]].sum().sort_values(by="Amount",ascending=False).reset_index()
plt.title("Age category- order relationship")
oo=sns.barplot(x='Age_category', y="Orders", data=aoa,width=0.3)
plt.figure(figsize=(10,5))
for bars in oo.containers:
    oo.bar_label(bars)
plt.show()
```

```python

plt.title("age-amount relationship")
aa=sns.barplot(x='Age_category', y="Amount", data=aoa,width=0.3)
for bars in aa.containers:
    aa.bar_label(bars)
```

```python
#Both orders and amount greatly done by middle aged persons 
```

```python
pro=df.groupby("Product_Category").agg({"Orders":"sum","Amount":"sum"}).sort_values(ascending=False, by="Amount")
print(pro)
```

```python
plt.title("product-amount relationship")
sca=sns.barplot(x="Product_Category",y="Amount",data=pro)
plt.xticks(rotation=45, ha="right")
for bars in sca.containers:
    sca.bar_label(bars, label_type="edge", padding=5, fmt="%.0f")
plt.show()
```

```python
plt.title("product-order relation")
pro22=df.groupby("Product_Category").agg({"Orders":"sum","Amount":"sum"}).sort_values(ascending=False, by="Orders")
ko=sns.barplot(x="Product_Category",y="Orders",data=pro22,palette="viridis", hue='Amount')
plt.xticks(rotation=45, ha="right")
for bars in ko.containers:
    ko.bar_label(bars)
plt.show()
```
